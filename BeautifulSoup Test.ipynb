```python
data_ref_id=[]
data_product_type=[]
data_product_name=[]
data_product_number=[]
data_price=[]
data_currency=[]
href=[]
name=[]
img=[]
```

```python
from IPython.display import HTML
```

```python
for product_data in product_wrappers:
#     print(product_data)
#     print("#############################")

# Store Main detail
    product_detail_obj=product_data.find(name='div', class_='range-revamp-product-compact')
    data_ref_id.append(int(product_detail_obj['data-ref-id']))
    data_product_type.append(product_detail_obj['data-product-type'])
    
# Store display name and link
    product_specific_namelink_obj=product_detail_obj.find(name='a')
    img.append(product_specific_namelink_obj.find('img',class_='range-revamp-aspect-ratio-image__image')['srcset'].split('\n')[10])
```
