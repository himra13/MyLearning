```python
def standard_clean(column_data):
    # Do nested chain to replace unique_id and proj_id
    return ((
    pd.Series(column_data)
    .str
    .replace('unique_id',key_map['unique_id'],regex=False)
    )
    .str
    .replace('proj_id',key_map['proj_id'],regex=False)
    .apply(lambda x: x.split(': ')))
```
